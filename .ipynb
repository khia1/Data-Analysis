```python

import pandas as pd
from IPython.display import display

stocks_path = r"/mnt/data/Stocks.csv"
prices_path = r"/mnt/data/Stock-Prices.csv"
```

## LOAD

```python

# 1) Load data and use 'Symbol' as the index. Display the DataFrame.
df = pd.read_csv(stocks_path)
assert 'Symbol' in df.columns, "Expected 'Symbol' column in Stocks.csv"
df = df.set_index('Symbol', drop=True)
display(df)
```

```python

# 2) Show how many non-null values exist for each column and each column's dtype.
print("=== df.info() (non-null counts and dtypes) ===")
df.info()
print("\n=== Summary table (non-null_count + dtype) ===")
summary = pd.DataFrame({
    "non_null_count": df.notna().sum(),
    "dtype": df.dtypes.astype(str)
})
display(summary)
```

```python

# 3) Use shape and len() to find rows and columns.
rows, cols = df.shape
print(f"Rows: {rows}, Columns: {cols}, len(df): {len(df)}, len(df.columns): {len(df.columns)}")
```

```python

# 4) Display the first and last 10 rows.
print("=== First 10 Rows ===")
display(df.head(10))
print("=== Last 10 Rows ===")
display(df.tail(10))
```

## ACCESS

```python

# 5) Show the Company column.
display(df['Company'].to_frame())
```

```python

# 6) Show the columns named Company and Industry.
display(df[['Company', 'Industry']])
```

```python

# 7) Present the information for the row that has the symbol AAPL.
if 'AAPL' in df.index:
    display(df.loc[['AAPL']])
else:
    print("AAPL not found in index.")
```

```python

# 8) Show the Exchange for the row with the Symbol AAPL.
if 'AAPL' in df.index and 'Exchange' in df.columns:
    print("Exchange for AAPL:", df.loc['AAPL', 'Exchange'])
elif 'AAPL' in df.index:
    print("Column 'Exchange' not found.")
else:
    print("AAPL not found in index.")
```

## FILTER

```python

# 9) Sort (and show) the data by Industry, then Company.
sorted_df = df.sort_values(by=['Industry', 'Company'], ascending=[True, True], kind='mergesort')
display(sorted_df)
```

```python

# 10) Show the rows where Industry is Information Technology.
it_rows = df[df['Industry'] == 'Information Technology']
display(it_rows)
```

## MISSING

```python

# 11) Show the rows where Notes has a value (not missing/empty).
if 'Notes' in df.columns:
    notes_present = df[df['Notes'].notna() & (df['Notes'].astype(str).str.strip() != "")]
    display(notes_present)
else:
    print("Column 'Notes' not found in Stocks.csv.")
```

## DROPPING

```python

# 12) Create and display a DataFrame excluding the Notes column.
if 'Notes' in df.columns:
    df_no_notes = df.drop(columns=['Notes'])
    display(df_no_notes)
else:
    df_no_notes = df.copy()
    print("Column 'Notes' not present; df_no_notes equals df.")
```

## JOIN

```python

# 13) Join with Stock-Prices.csv (Symbol, Price) using join()
prices = pd.read_csv(prices_path)
assert 'Symbol' in prices.columns and 'Price' in prices.columns, "Expected 'Symbol' and 'Price' in Stock-Prices.csv"
prices = prices.set_index('Symbol', drop=True)
df_with_prices = df.join(prices[['Price']], how='left')
display(df_with_prices)

# Save the joined output for submission
output_path = "/mnt/data/Stocks_with_Prices.csv"
df_with_prices.to_csv(output_path, index=True)
print("Saved:", output_path)
```
